```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
df = pd.read_csv('../data/raw/Pimadiabetes.csv')

# 2. Basic Inspection
print("Dataset Shape:", df.shape)
print("\nFirst 5 Rows:")
print(df.head())

# 3. Identify Hidden Zeros
# Columns where 0 is medically impossible
cols_to_fix = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

print("\nZero counts before fixing:")
for col in cols_to_fix:
    print(f"{col}: {(df[col] == 0).sum()} zeros")

# 4. Replace 0 with NaN (so they don't mess up the Median calculation)
df[cols_to_fix] = df[cols_to_fix].replace(0, np.nan)

# 5. Fill NaN with Median (Standard practice for this dataset)
df.fillna(df.median(), inplace=True)

# 6. Save the cleaned version
df.to_csv('../data/processed/Pima_Cleaned.csv', index=False)
print("\nCleaned data saved to data/processed/Pima_Cleaned.csv")

# 7. Visualization: Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation with Diabetes Outcome")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Correlation Heatmap (Shows how features relate to diabetes)
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='RdYlGn', fmt='.2f')
plt.title('Clinical Feature Correlation Heatmap')
plt.show()

# 2. Glucose Distribution by Outcome
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='Glucose', hue='Outcome', fill=True)
plt.title('Glucose Distribution: Healthy vs Diabetic')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CORRECT PATH based on your folder structure
# We go up one level (../), then into data/raw/
df = pd.read_csv('../data/raw/Pimadiabetes.csv') 

# Check if data loaded
print("Data Shape:", df.shape)

# Correlation Heatmap
plt.figure(figsize=(10, 8))
# We use numeric_only=True to avoid errors if there are text columns
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='RdYlGn', fmt='.2f')
plt.title('Clinical Feature Correlation Heatmap')
plt.show()

# Glucose Distribution
plt.figure(figsize=(8, 5))
# Note: Ensure the target column name matches your CSV (usually 'Outcome' or 'Diabetes')
sns.kdeplot(data=df, x='Glucose', hue='Outcome', fill=True) 
plt.title('Glucose Distribution: Healthy vs Diabetic')
plt.show()
```
